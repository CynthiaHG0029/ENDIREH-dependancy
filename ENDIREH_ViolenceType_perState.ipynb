```python
correl = pd.read_csv('./correl.csv', encoding='latin-1')
```

```python
correl1 = correl.melt(id_vars=['Situación laboral',	'Factor de expansión de la vivienda','Factor de expansión de las mujeres elegidas','Identificador de la mujer'])
```

```python
correl2 = (correl1[(correl1.value == 1.0)])
```

```python
correl2
```

```python
correl3 = correl2.pivot(index=['Situación laboral',	'Factor de expansión de la vivienda','Factor de expansión de las mujeres elegidas','Identificador de la mujer'],columns='variable',values='value' )
```

```python
correl3.reset_index()
```

```python
entidad = pd.read_csv('./entidad.csv')
```

```python
entidad
```

```python
mapa = correl3.merge(entidad, how='left', on='Identificador de la mujer')
```

```python
mapa.columns.tolist()
```
